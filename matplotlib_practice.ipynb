```python
%matplotlib inline
```

```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create a graph with subplot
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add lable
plt.plot(x_axis, y_axis, color = "Green", linewidth = 2, label = 'Boston', marker='D')
# Create label for x adnd y axis
plt.xlabel("Date")
plt.ylabel("Fair($)")

# Marker style
fmt = '[D]'

# Set y limit betwin 0-45
plt.ylim(0, 45)

# Create a title
plt.title("PyBer Fare by Month")

# Add the legend
plt.legend()

# Add grid
plt.grid()
```

```python
# Create plot
plt.bar(x_axis,y_axis)
```

```python
# Create plot
plt.bar(x_axis, y_axis, color = 'green', label = 'Boston')

plt.xlabel('Months')
plt.ylabel('Fare($)')
plt.ylim(0,45)
plt.legend()
plt.title("PyBer Fare by Month")
plt.grid()
```

```python
plt.barh(x_axis,y_axis)
```

```python
plt.barh(x_axis, y_axis, color = "Magenta", label = "Boston")
plt.gca().invert_yaxis()
plt.xlabel('Fare($)')
plt.ylabel('Month')
plt.title("PyBer Fare by Month")
plt.legend()
plt.grid()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(y_axis, x_axis, color = "cyan", label = "Chicago")
ax.barh().invert_yaxis()
```
